```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### load the cancer dataset

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer
```

```python
cancer.keys()
```

```python
print(cancer.DESCR)
```

```python
df = pd.DataFrame(cancer["data"], columns= cancer["feature_names"])
df
```

#### Number of dimensions are 30  and we have to reduce it to 2

#### We can use PCA to find the first two principal components and visualise the data

#### before we do this, we need to scale the data so that each feature has a single unit variance

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scalar.fit(df)
```

```python
scaled_data = scalar.transform(df)
scaled_data.shape
```

```python
def plots(df, col, model):
    plt.figure(figsize=(13,5))
    plt.subplot(121)
    sns.kdeplot(df[col])
    plt.title('before stdscalar')
    
    plt.subplot(122)
    p1 = model.fit_transform(df[[col]]).flatten()
    sns.kdeplot(p1)
    plt.title('after stdscalar')
```

```python
for col in df.columns:
    plots(df, col, StandardScaler())
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2) #we have to reduce the 30 features to 2 features
pca.fit(scaled_data) 
```

```python
x_pca =pca.transform(scaled_data)
x_pca.shape
```

#### we have reduced the 30 features to 2 features

#### lets plot it on the graph

```python
plt.figure(figsize= (15,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=cancer["target"], cmap= "plasma")
plt.xlabel("First Principle Component")
plt.ylabel("Second Principle Component")
```

#### By using the first two important components we can seperate the classes
